```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
file = "./app_table.csv"
apps_df = pd.read_csv(file)
apps_df
```

```python
new_df = apps_df.rename(columns={"index": "id",
                                                          "app": "App", 
                                                          "price": "Price",
                                                           "reviews":"Total_Ratings",
                                                            "genres":"Genre", "rating":"User_Rating", "os":"OS"})
new_df.head()
```

```python
new_df.set_index("id", inplace=True)
```
