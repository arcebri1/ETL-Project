```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
apple_file = "AppleStore.csv"
apple_df = pd.read_csv(apple_file)
apple_df.head()
```

```python
new_apple_df=apple_df[['track_name', 'price', 'rating_count_tot', 'prime_genre', 'user_rating' ]].copy()
new_apple_df.head() 
```

```python
apple_transf = new_apple_df.rename(columns={"track_name": "App",
                                                          "price": "Price",
                                                          "rating_count_tot": "Reviews",
                                                           "prime_genre":"Genres",
                                                            "user_rating":"Rating"})
apple_transf.head()
```

```python
google_file = "googleplaystore.csv"
google_df = pd.read_csv(google_file)
google_df.head()
```

```python
new_google_df=google_df[['App', 'Price', 'Reviews','Genres', 'Rating' ]].copy()
new_google_df.head() 
```
